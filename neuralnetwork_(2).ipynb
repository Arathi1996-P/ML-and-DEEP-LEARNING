```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Dropout
import warnings
warnings.filterwarnings('ignore')
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
df = pd.read_csv(io.BytesIO(uploaded['housepricedata.csv']))
df
```

```python
x = df.drop('AboveMedianPrice',axis=1).values
y = df.AboveMedianPrice
```

```python
from sklearn.model_selection import train_test_split
x_train, x_val_and_test, y_train,y_val_and_test = train_test_split(x,y, test_size=0.20)
x_val, x_test, y_val, y_test = train_test_split(x_val_and_test, y_val_and_test, test_size=0.5)
```

```python
print(x_train.shape, x_val.shape, x_test.shape, y_train.shape, y_val.shape, y_test.shape)
```

```python
model = Sequential([
    Dense(30, activation='relu', input_shape=(10,)),
    Dropout(0.2),
    Dense(30, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid'),
    Dropout(0.2)
])
```

```python
model.compile(optimizer='sgd',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train, y_train,batch_size=32, epochs=100,validation_data=(x_val, y_val))
```

```python
model.evaluate(x_test, y_test)
```

```python
model2 = Sequential([
    Dense(100, activation='relu', input_shape=(10,)),
    Dense(100, activation='relu'),
    Dense(100, activation='relu'),
    Dense(100, activation='relu'),
    Dense(1, activation='sigmoid'),
])
```

```python
model2.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
hist2 = model2.fit(x_train, y_train,batch_size=32, epochs=100,validation_data=(x_val, y_val))
```

```python
model2.evaluate(x_test, y_test)
```

```python
prediction = model2.predict([x_test])
print(prediction)
```

plot loss and accuracy

```python
plt.plot(hist2.history['loss'])
plt.plot(hist2.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'])
plt.show()
```

```python
plt.plot(hist2.history['accuracy'])
plt.plot(hist2.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='lower right')
plt.show()
```
